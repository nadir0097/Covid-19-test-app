```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
covid = pd.read_csv("covid_19_complete.csv")

print(covid.describe(include="all"))

print(covid.loc[covid["Country/Region"]=="US",["Confirmed","Deaths","Recovres"]])
print(covid.loc[covid["Country/Region"]=="Algeria",["Confirmed","Deaths","Recovres"]])

data=covid.loc[covid["Country/Region"].isin(["Algeria","US"]),:]
plt.figure(figsize=(15,8))
sns.lineplot(data=data,x="Date",y="Confirmed",hue="Country/Region")

```
